```python
import pandas as pd
```

```python
# Reading the data from the cvs file
game_df = pd.read_csv("Resources/purchase_data.csv")
```

```python
# Looking into the frame to understand the information
game_df.head()
```

```python
# Look at the type of value in the columns
game_df.dtypes
```

```python
# Looking for posible lost values in the data to clean it
game_df.count()
```

```python
# Everything looks right to continue to the study!
# PLAYER COUNT
# Total number of player, one player can make more than one purchase
total_player = len(game_df['SN'].unique())
pd.DataFrame({"Total Players": [total_player]})
```

```python
# PURCHASING ANALYSIS
# Number of unique items
total_unique_items = len(game_df['Item Name'].unique())
# Average purchased price
average_purchase = round(game_df['Price'].mean(), 2)
# Total number of purchases
total_purchases = game_df['Purchase ID'].count()
# Total revenue
revenue = game_df['Price'].sum()
# Creating a frame for the results
purchasing = pd.DataFrame({"Number of Unique Items" : [total_unique_items],
                          "Average Price" : [average_purchase],
                          "Number of Purchases" : [total_purchases],
                          "Total Revenue" : [revenue]})
purchasing['Average Price'] = purchasing['Average Price'].map("${:.2f}".format)
purchasing['Total Revenue'] = purchasing['Total Revenue'].map("${:,}".format)
purchasing
```

```python
# GENDER DEMOGRAPHIC
# Percentage and count of male player
# Creating the group by objet from game_df by gender
purchasing_gender = game_df.groupby("Gender")
counts_gender = purchasing_gender.SN.nunique()
percentage_by_gender = [round(value/total_player*100, 2) for value in counts_gender]
# Placing solution into a Frame
gender_demographic = pd.DataFrame()
gender_demographic['Total Counts'] = counts_gender
gender_demographic['Percentage of Players'] = percentage_by_gender
gender_demographic['Percentage of Players'] = gender_demographic['Percentage of Players'].map("{:.2f}%".format)
gender_demographic.sort_values("Total Counts",ascending=False)
gender_demographic
```

```python
# PURCHASING GENDER ANALYSIS
pg_count = purchasing_gender['Purchase ID'].count()
pg_mean = purchasing_gender['Price'].mean()
pg_sum = purchasing_gender['Price'].sum()
# We must calculate the number of clients that are in each gender to calculate the average 
# Total Purchase per Person, dividing the total sum per gender by the number of clients per gender
number_clients_per_gender = [len(item) for item in purchasing_gender.SN.unique()]
purchase_gender = pd.DataFrame({"Purchased Count": pg_count,
                                "Average Purchase Price" : pg_mean,
                                "Total Purchase Value" : pg_sum,
                                "Avg Total Purchase per Person" : pg_sum/number_clients_per_gender })
# Changing the format in the Data Frame to print
purchase_gender['Average Purchase Price'] =  purchase_gender['Average Purchase Price'].map("${:.2f}".format)
purchase_gender['Total Purchase Value'] =  purchase_gender['Total Purchase Value'].map("${:,.2f}".format)
purchase_gender['Avg Total Purchase per Person'] =  purchase_gender['Avg Total Purchase per Person'].map("${:.2f}".format)
purchase_gender
```

```python
# AGE DEMOGRAPHICS
# Creating a bin that will help split the data frame
bins = [0,9,14,19,24,29,34,39,100]
# Placing the bins' name
bins_name = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
game_df['Age Group'] = pd.cut(game_df['Age'], bins, labels=bins_name, include_lowest = True)
# Create the groupby object based on 'Age Group'
purchase_age = game_df.groupby("Age Group")
# Saving the results to variables
# Counting the unique users
age_counts = purchase_age.SN.nunique()
age_percentage_count = [ round(100*item/total_player, 2) for item in age_counts]
# Place the solution into a frame
age_demographic = pd.DataFrame()
age_demographic['Total Counts'] = age_counts
age_demographic['Percentage of Players'] = age_percentage_count
age_demographic['Percentage of Players'] = age_demographic['Percentage of Players'].map("{:.2f}%".format)
age_demographic
```

```python
# PURCHASING AGE ANALYSIS
# Purchasing Count
pa_count = purchase_age.SN.count()
# Average Purchase Price
pa_average = purchase_age.Price.mean()
# Total Purchase Value
pa_total = purchase_age.Price.sum()
# Average Purchase by Person
# age_counts has in it the total user divided by age's bin
pa_avg_by_person = pa_total/age_counts
# Saving the results into a frame
pa_frame = pd.DataFrame({ "Purchase Count":pa_count, "Average Purchase Price": pa_average,
                            "Total Purchase Value":pa_total, "Avg Purchase by Person": pa_avg_by_person})
pa_frame['Average Purchase Price'] = pa_frame['Average Purchase Price'].map("${:.2f}".format)
pa_frame['Total Purchase Value'] = pa_frame['Total Purchase Value'].map("${:.2f}".format)
pa_frame['Avg Purchase by Person'] = pa_frame['Avg Purchase by Person'].map("${:.2f}".format)
pa_frame
```

```python
# TOP SPENDERS
# Creating a group_by using SN
sn_frame = game_df.groupby("SN")
# Purchase Count
snf_count = sn_frame.SN.count()
# Average Purchase Price
snf_avg_purchase = sn_frame.Price.mean()
# Total Purchase Value
snf_total_purchase = sn_frame.Price.sum()
# Placing the results into a frame
snf = pd.DataFrame({"Purchase Count": snf_count, "Average Purchase Price": snf_avg_purchase,
                    "Total Purchase Value" : snf_total_purchase})
snf["Average Purchase Price"] = snf["Average Purchase Price"].map("${:.2f}".format)
# Line to give format to the Total Purchase Value
#  (*) snf["Total Purchase Value"] = snf["Total Purchase Value"].map("${:.2f}".format)
# Sorting the values from Total Purchase Value - descending
snf.sort_values("Total Purchase Value", ascending=False).head(5)
```

```python
# MOST POPULAR BY PURCHASE COUNT
# Creating a groupby object using 'Item Name'
mppc = game_df.groupby(["Item ID","Item Name"])
mppc_count = mppc["Purchase ID"].count()

mppc_item_price= mppc.Price.unique()

mppc_total_purchase = mppc["Price"].sum()
mppc_frame = pd.DataFrame({"Purchase Count":mppc_count,"Item Price":mppc_item_price,
                           "Total Purchase Value":mppc_total_purchase})
mppc_frame.sort_values('Purchase Count', ascending=False).head(5)

#mppc_item_price
```
